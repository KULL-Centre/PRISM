## Find overlapping sequences

```python
uniprot_accession= 'P60484'
df_dic[uniprot_accession]

```

```python
pdb_dic={}
for n in df_dic:
    
    final,coverage,add=findbest(df_dic[n],n)
    pdb_dic[n]=[final,coverage,add]
    try:
        if final != []:
            print('UniprotID: ',n,'\nChoosen_pair:  ',final[0],'\nCoverage of possible: ',f'{round(final[2]/coverage*100,1)}%','\nCoverage of uniprot: ',f'{round(final[2]/len(add)*100,1)}%','\nOverlaps: ',final[1],'\nResolution: ',final[3])
            pdbid_list=final[0].split('_')
            if pdbid_list[0]==pdbid_list[1]:
                pdbid_list=[pdbid_list[0]]
            plot_pdb_seq(df_dic[n],n,pdbid_list)
            plt.suptitle(f'{n}\nCover of possible: {round(final[2]/coverage*100,1)}% - Cover of uni: {round(final[2]/len(add)*100,1)}% - Resolution: {final[3]} - Overlaps: {final[1]}', fontsize=14)
            print('---------------------------------------')
        else:
            print(f'Error for {n} due to no combinations within 95% cover range')
            print('---------------------------------------')
    except:
        print(f'Error for {n}')
        print('---------------------------------------')
        
```
